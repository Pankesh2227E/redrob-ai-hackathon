```python
import json
import pandas as pd

# -----------------------------
# Paths
# -----------------------------
CANDIDATES_PATH = "challenge/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl"

# -----------------------------
# Keywords
# -----------------------------

HIGH_VALUE_SKILLS = {
    "python","pytorch","tensorflow","llm","lora","qlora",
    "peft","embedding","embeddings","milvus","pinecone",
    "weaviate","qdrant","faiss","elasticsearch","opensearch",
    "retrieval","rag","nlp","transformers","bert","sql",
    "spark","airflow","aws","gcp","azure"
}

GOOD_TITLES = {
    "ai engineer","machine learning engineer","ml engineer",
    "backend engineer","software engineer","data engineer",
    "search engineer","recommendation engineer",
    "applied scientist"
}

BAD_TITLES = {
    "marketing","sales","hr","content writer",
    "graphic designer","recruiter"
}

GOOD_COMPANIES = {
    "google","microsoft","amazon","flipkart","swiggy",
    "zomato","razorpay","phonepe","meesho","uber",
    "atlassian","adobe","netflix","linkedin"
}

CONSULTING = {
    "tcs","infosys","wipro","accenture","capgemini",
    "cognizant","mindtree","hcl","tech mahindra"
}

# -----------------------------
# Scoring
# -----------------------------

def score_candidate(c):

    score = 0
    reasons = []

    profile = c.get("profile", {})
    career = c.get("career_history", [])
    skills = c.get("skills", [])
    signals = c.get("redrob_signals", {})

    # ---------------------
    # Experience
    # ---------------------

    yrs = profile.get("years_of_experience",0)

    if 5 <= yrs <= 9:
        score += 15
        reasons.append(f"{yrs:.1f} yrs exp")
    elif yrs > 9:
        score += 10
    elif yrs >=3:
        score += 5

    # ---------------------
    # Title
    # ---------------------

    title = profile.get("current_title","").lower()

    for t in GOOD_TITLES:
        if t in title:
            score += 15
            reasons.append(title)
            break

    for t in BAD_TITLES:
        if t in title:
            score -= 20

    # ---------------------
    # Summary
    # ---------------------

    summary = profile.get("summary","").lower()

    important_words = [
        "retrieval","ranking","recommendation",
        "embedding","llm","python",
        "search","backend","spark"
    ]

    for w in important_words:
        if w in summary:
            score += 3

    # ---------------------
    # Skills
    # ---------------------

    ai_skill_count = 0

    for s in skills:

        name = s["name"].lower()

        if name in HIGH_VALUE_SKILLS:
            ai_skill_count += 1
            score += 4

            if s.get("proficiency","") == "advanced":
                score += 2

    reasons.append(f"{ai_skill_count} AI skills")

    # ---------------------
    # Career
    # ---------------------

    for job in career:

        company = job.get("company","").lower()

        desc = job.get("description","").lower()

        title = job.get("title","").lower()

        if any(x in company for x in GOOD_COMPANIES):
            score += 10

        if any(x in company for x in CONSULTING):
            score -= 6

        keywords = [
            "retrieval","ranking","recommendation",
            "search","embedding","vector",
            "backend","python","spark",
            "machine learning","ml"
        ]

        for k in keywords:
            if k in desc:
                score += 2

        if "backend" in title:
            score += 4

        if "ml" in title or "machine learning" in title:
            score += 6

    # ---------------------
    # Behaviour
    # ---------------------

    if signals.get("open_to_work_flag"):
        score += 5

    if signals.get("notice_period_days",90) <= 30:
        score += 5

    score += min(signals.get("github_activity_score",0),10)

    score += min(signals.get("profile_completeness_score",0)/20,5)

    score += signals.get("recruiter_response_rate",0)*10

    # ---------------------
    # Final
    # ---------------------

    return score, "; ".join(reasons)


# -----------------------------
# Read all candidates
# -----------------------------

results = []

with open(CANDIDATES_PATH,"r") as f:

    for line in f:

        c = json.loads(line)

        score, reason = score_candidate(c)

        results.append({
            "candidate_id": c["candidate_id"],
            "score_raw": score,
            "reasoning": reason
        })

# -----------------------------
# Ranking
# -----------------------------

df = pd.DataFrame(results)

df = df.sort_values(
    "score_raw",
    ascending=False
).reset_index(drop=True)

df = df.head(100)

# normalize score 0-1

mx = df["score_raw"].max()
mn = df["score_raw"].min()

df["score"] = ((df["score_raw"]-mn)/(mx-mn+1e-6)).round(3)

df["rank"] = range(1,len(df)+1)

submission = df[[
    "candidate_id",
    "rank",
    "score",
    "reasoning"
]]

submission.to_csv("submission.csv",index=False)

print(submission.head())

print("\nSaved as submission.csv")
```

```python
import json
import pandas as pd
from sentence_transformers import util

CANDIDATES_PATH = "challenge/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl"

results = []

with open(CANDIDATES_PATH) as f:
    for i, line in enumerate(f):

        candidate = json.loads(line)

        # Build candidate text
        profile = candidate["profile"]

        career_text = " ".join(
            job.get("title","") + " " + job.get("description","")
            for job in candidate.get("career_history", [])
        )

        skills_text = " ".join(
            skill.get("name","")
            for skill in candidate.get("skills", [])
        )

        candidate_text = f"""
        {profile.get("headline","")}
        {profile.get("summary","")}
        {career_text}
        {skills_text}
        """

        # Semantic similarity
        candidate_embedding = model.encode(
            candidate_text,
            convert_to_tensor=True,
            normalize_embeddings=True
        )

        semantic = float(
            util.cos_sim(job_embedding, candidate_embedding)
        )

        # Structured score
        structure = structured_score(candidate)

        # Normalize structure score (approximate)
        structure = min(structure / 100.0, 1.0)

        # Final score
        final_score = 0.65 * semantic + 0.35 * structure

        results.append({
            "candidate_id": candidate["candidate_id"],
            "semantic": semantic,
            "structure": structure,
            "score": final_score
        })

        # Progress
        if (i + 1) % 1000 == 0:
            print(f"Processed {i+1} candidates")

        # TEMPORARY LIMIT
        if i == 4999:
            break

df = pd.DataFrame(results)

df = df.sort_values(
    "score",
    ascending=False
)

df.head(20)
```

```python
AI_KEYWORDS = [
    "python",
    "machine learning",
    "deep learning",
    "llm",
    "lora",
    "qlora",
    "fine",
    "embedding",
    "retrieval",
    "vector",
    "milvus",
    "pinecone",
    "weaviate",
    "qdrant",
    "faiss",
    "elasticsearch",
    "opensearch",
    "nlp",
    "transformer",
    "huggingface",
    "pytorch",
    "tensorflow",
    "spark",
    "airflow",
    "kafka",
    "sql",
    "aws",
    "azure",
    "gcp",
    "docker",
    "kubernetes",
    "xgboost",
    "recommendation"
]

def skill_score(candidate):
    matched = []

    for skill in candidate["skills"]:
        name = skill["name"].lower()

        for keyword in AI_KEYWORDS:
            if keyword in name:
                matched.append(skill["name"])
                break

    return len(matched), matched
```

```python
def structured_score(candidate):
    score = 0

    # Experience (JD prefers 5–9 years)
    exp = candidate["profile"].get("years_of_experience", 0)

    if 5 <= exp <= 9:
        score += 10
    elif 4 <= exp <= 10:
        score += 7
    else:
        score += 3

    # Open to work
    if candidate["redrob_signals"]["open_to_work_flag"]:
        score += 5

    # Notice period
    notice = candidate["redrob_signals"]["notice_period_days"]

    if notice <= 30:
        score += 5
    elif notice <= 60:
        score += 3

    # Recruiter response rate
    score += candidate["redrob_signals"]["recruiter_response_rate"] * 5

    # GitHub activity
    score += candidate["redrob_signals"]["github_activity_score"] / 2

    return score
```

```python
def final_score(candidate, semantic_score):
    structured = structured_score(candidate)

    # 70% semantic + 30% structured
    return (
        semantic_score * 0.7 +
        (structured / 40) * 0.3
    )
```

```python
def create_candidate_text(candidate):
    profile = candidate["profile"]

    career_text = " ".join(
        f"{job['title']} {job['description']}"
        for job in candidate["career_history"]
    )

    skills_text = " ".join(
        skill["name"]
        for skill in candidate["skills"]
    )

    return f"""
Headline:
{profile['headline']}

Summary:
{profile['summary']}

Career:
{career_text}

Skills:
{skills_text}
"""

def structured_score(candidate):
    score = 0

    exp = candidate["profile"].get("years_of_experience", 0)

    if 5 <= exp <= 9:
        score += 10
    elif 4 <= exp <= 10:
        score += 7
    else:
        score += 3

    signals = candidate["redrob_signals"]

    if signals["open_to_work_flag"]:
        score += 5

    notice = signals["notice_period_days"]

    if notice <= 30:
        score += 5
    elif notice <= 60:
        score += 3

    score += signals["recruiter_response_rate"] * 5
    score += signals["github_activity_score"] / 2

    return score


def final_score(candidate, semantic_score):
    structured = structured_score(candidate)
    return semantic_score * 0.7 + (structured / 40) * 0.3
```
